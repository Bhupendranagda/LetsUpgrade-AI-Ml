```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Loading the attrition dataset

```python
df = pd.read_csv('general_data.csv')
```

```python
df.head()
```

# Performing datacleaning and preprocesing

```python
df.isna().sum()
```

```python
mode_NumOfComp = df['NumCompaniesWorked'].mode()
new_NumOfComp = np.where(df['NumCompaniesWorked'].isnull(),mode_NumOfComp,df['NumCompaniesWorked'])
```

```python
meadian_TotalWorking = df['TotalWorkingYears'].median()
new_totWor = np.where(df['TotalWorkingYears'].isnull(),meadian_TotalWorking,df['TotalWorkingYears'])
```

```python
df['NumCompaniesWorked'] = new_NumOfComp
df['TotalWorkingYears'] = new_totWor
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
#Convert categorical to numeric
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Attrition'] = le.fit_transform(df['Attrition'])
```

```python
df.info()
```

```python
df['BusinessTravel'] = le.fit_transform(df['BusinessTravel'])
df['Department'] = le.fit_transform(df['Department'])
df['Gender'] = le.fit_transform(df['Gender'])
df['JobRole'] = le.fit_transform(df['JobRole'])
df['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
df['Over18'] = le.fit_transform(df['Over18'])
df['EducationField'] = le.fit_transform(df['EducationField'])

```

```python
df.head()
```

# RandomForest for Important Variable

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df.head()
```

```python
rf_model = RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features = ['Age','BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EmployeeCount', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
rf_model.fit(X=df[features],y=df['Attrition'])
```

# OOB Score and Imp Variable Selection

```python
print("OBB Score ")
print(rf_model.oob_score_)
```

```python
for feature,imp in zip (features,rf_model.feature_importances_):
    print(feature,imp)
```

## All important Variables are
* And out of them most important are Age and Monthlyincome

```python
for feature,imp in zip (features,rf_model.feature_importances_):
    if imp>0.07:
        print(feature,imp)
```
